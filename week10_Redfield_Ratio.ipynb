## 1. Download the bottle data in CSV format
(https://cchdo.ucsd.edu/data/14186/33RO20161119_hy1.csv). Read in the bottle data into MATLAB or Python. Use the GSW package to calculate absolute salinity (hint: gsw_SA_from_SP) and conservative temperature (hint: gsw_CT_from_t).

```python
# Step 1 download data from BATS and create data vectors
df=pd.read_csv('33RO20161119_hy1.csv',skiprows=149)
pd.set_option('display.max_columns', 500)
df
```

```python
# get data so we can calculate sigma-theta
pres=df['CTDPRS'][1:-1].astype(float)
sal=df['CTDSAL'][1:-1].astype(float)
temp=df['CTDTMP'][1:-1].astype(float)
lon=df['LONGITUDE'][1:-1].astype(float)
lat=df['LATITUDE'][1:-1].astype(float)
sflg=df['CTDSAL_FLAG_W'][1:-1].astype(int)
```

```python
# first calculate absolute salinity
sa=gsw.SA_from_SP(sal,pres,lon,lat)
# then calculate conservative temperature
ct=gsw.CT_from_t(sa,temp,pres)
# then sigma theta
st=gsw.sigma0(sa,ct)
```

```python
# now load N and P
no3=df['NITRAT'][1:-1].astype(float)
po4=df['PHSPHT'][1:-1].astype(float)
o2=df['OXYGEN'][1:-1].astype(float)
nflg=df['NITRAT_FLAG_W'][1:-1].astype(int)
pflg=df['PHSPHT_FLAG_W'][1:-1].astype(int)
```

```python
# define conditions to include data
isgood=(nflg==2)&(pflg==2)
oxic=(o2>=20)
# combine conditions
ind=(isgood==True)&(oxic==True) 
```

```python
# trim down the data that matches desired conditions
x=po4[ind]
y=no3[ind]
```

```python
# perform linear regression
c=np.cov(x,y)
a=c[0,1]/c[0,0] # slope
b=np.mean(y)-a*np.mean(x) # intercept
# calculate R2 value
r2=c[0,1]**2/c[0,0]/c[1,1]
```

```python
# expand the regression line
x0=np.linspace(0,3,30)
y0=a*x0+b

# plot the result
plt.plot(x,y,'.')
plt.plot(x0,y0,'r-')
plt.xlabel('phosphate, umol/kg')
plt.ylabel('nitrate, umol/kg')
print('N:P ratio is '+str(a.round(2))+' and R2 value is '+str(r2.round(2)))
```
